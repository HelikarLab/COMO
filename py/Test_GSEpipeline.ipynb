```python
# Calculate Z score
def calculate_z_score(df):
    cols = list(df)
    result = pd.DataFrame([],index=df.index)
    for col in cols:
        if not '.gz.' in col:
            score = np.log2(df[col])
            result[col] = (score-score.mean())/score.std(ddof=0)
    return result
```

```python
df_zscore = calculate_z_score(df_clean)
df_zscore
```
