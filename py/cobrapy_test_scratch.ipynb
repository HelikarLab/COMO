````python
model_opt = model.optimize().to_frame()

fluxsolution = pd.DataFrame()
for i in range(len(model.genes)):
    if not model.genes[i].id in DT_model:
        continue
    model.genes[i].knock_out()
    opt_model = model.optimize().to_frame()
    fluxsolution[model.genes[i].id]=opt_model['fluxes']

fluxsolution
````

````python
fluxSolutionRatios = fluxsolution.div(model_opt['fluxes'],axis=0)
fluxSolutionRatios
fluxSolutionDiffs = fluxsolution.sub(model_opt['fluxes'],axis=0)
fluxSolutionDiffs
````

````python
HasEffects_Gene = []
for i in range(len(model.genes)):
    geneid = model.genes[i].id
    if not geneid in DT_model:
        continue
    for rxn in model.genes[i].reactions:
        if fluxSolutionDiffs.at[rxn.id,geneid] != 0:
            HasEffects_Gene.append(geneid)
            break
HasEffects_Gene
````
